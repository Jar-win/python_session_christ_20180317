```python
import pandas as pd
import seaborn as sns
import numpy as np

from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Pokemon.csv', index_col=0)
```

```python
df.head()
```

```python
sns.lmplot(x='Attack', y='Defense', data=df)
```

```python
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=True, # No regression line
           hue='Stage')
```

```python
# Plot using Seaborn
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=False, 
           hue='Stage')
 
# Tweak using Matplotlib
plt.ylim(0, None)
plt.xlim(0, None)
```

```python
sns.boxplot(data=df)
```

```python
corr = df.corr() # correlation
```

```python
corr.head()
```

#### Color matrix using matplotlib

```python
plt.figure(figsize=(10,10))
plt.pcolor(corr)
plt.colorbar()
plt.show()
```

#### Color Matrix using Seaborn

```python
sns.heatmap(corr)
```

```python
sns.distplot(df.Attack)
```
